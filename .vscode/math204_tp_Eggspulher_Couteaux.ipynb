3. Soit les matrices B := $\begin{bmatrix}1 & 0 &-1 \\ 1 &1 & 0\end{bmatrix}$ et C := $\begin{bmatrix}1 & 1 \\ 1 &0 \\ 1 & 1\end{bmatrix}$

Le produit B*C donne : $\begin{bmatrix}0 & 0 \\ 2 &1\end{bmatrix}$

Le produit C*B donne : $\begin{bmatrix}2 & 1 & -1 \\ 1 & 0 & -1 \\ 2 & 1 & -1\end{bmatrix}$

4. D'abord, il faut vérifier que le produit des deux matrices est possible.

Le nombre de colonnes de la première matrice, doit être égal au nombre de lignes de la seconde.

Si A est de taille (m * n) et B de taille (n * p), la matrice résultante C sera de taille (m * p).

Pour obtenir l'élément situé à la ligne i et à la colonne j de la matrice C, on prend la ligne i de la matrice A, et on effectue le produit scalaire avec la colonne j de la matrice B.


La produit matriciel n'est pas commutatif.
En effet : $A \times B \neq B \times A$



```python
def produitdeM(m1, m2):
    """
    Produit matriciel de deux matrices Sage.
    """
    if m1.ncols() != m2.nrows() :
        return None
    C = matrix(m1.nrows(), m2.ncols())
    for i in range(0,m1.nrows()):
        for j in range(0,m2.ncols()):
            ligne_i = m1.row(i)
            col_j = m2.column(j)
            C[i, j] = ligne_i * col_j 
    return C

produitdeM(B, C) # exemple
    
```

3. Pour savoir si une famille est libre, on regarde si son rang est égal à son nombre de colonne.

```python
generatrice = M.rank() == 3
generatrice
```

Une famille est génératrice si le rang de la matrice est égal à la dimension de l'espace vectoriel.

```python
libre and generatrice
```

Une famille est une base si elle est libre et génératrice.

```python
def est_base(M : Matrix, dim = 3) -> bool:
    """
    DIMENSION 3 par défaut.
    """
    libre = M.rank() == M.nrows()
    generatrice = M.rank() == dim
    return libre and generatrice
```
